```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from scipy.fftpack import fft
import math
# if using a jupyter notebook
%matplotlib inline  
```

<div dir=rtl>
ابتدا توابع مورد نیاز برای رسم نمودار هارا مینویسم، تابع اول تابع ها را بصورت گسسته رسم می‌کند و دومی توابع را بصورت پیوسته رسم میکند.
</div>

```python
def stem(data):
  x = data[0]
  y = data[1]
  plt.stem(x, y)
  plt.show()
def plot(data):
  x = data[0]
  y = data[1]
  plt.plot(x, y)
  plt.show()
```

<div dir=rtl>
الف)

در این بخش تابع تولید کننده سیگنال سینوس طبق متن تمرین را پیاده سازی میکنم،  برای بازه داده شده سینوس نقاط را پیدا میکنم و بصورت یک آرایه مختصات x و y را برمی‌گردانم.
</div>

```python
def sin_func(N, M):
  x = []
  y = []
  omega = 2*np.pi*M/N
  for i in range(N):
    tmp_x = omega*i
    x.append(tmp_x)
    y.append(math.sin(tmp_x))
  return x, y
```

<div dir=rtl>
الف - ادامه)
در این بخش تابع تولید کننده سیگنال کسینوس طبق متن تمرین را پیاده سازی میکنم،  برای بازه داده شده سینوس نقاط را پیدا میکنم وبصورت یک آرایه مختصات x و y را برمی‌گردانم.
</div>

```python
def cos_func(N, M):
  x = []
  y = []
  omega = 2*np.pi*M/N
  for i in range(N):
    tmp_x = omega*i
    x.append(tmp_x)
    y.append(math.cos(tmp_x))
  return x, y
```

<div dir=rtl>
در این بخش تابعی را نوشتم که با گرفتن دو تابع و دو ضریب، ترکیب خطی آنهارا برمیگرداند:
</div>

```python
def sum(func_1, func_2, a = 1, b = 1):
  x_1 = func_1[0]
  x_2 = func_2[0]
  y_1 = func_1[1]
  y_2 = func_2[1]

  x = []
  y = []

  for i in range(len(x_1)):
    x.append(x_1[i])
    y.append(a*y_1[i] + b*y_2[i])

  return x, y
```

<div dir=rtl>
ب)
در این بخش تابع پالس مربعی را پیاده‌سازی کردم، توجه کنید این تابع در مرز ها مقدار 1/2 را برمی‌گرداند.
</div>

```python
def rect_func(N):
  x = []
  y = []
  for i in range(N):
    x.append(i)
    if i < N/4 or i >= 3*N/4:
      y.append(0)
    elif i >= N/4 and i < 3*N/4:
      y.append(1)
    else:
      y.append(1/2)
  return x, y
```

```python
def ft_top_n_component(x, n):
  fourier = np.fft.fft(np.array(x[1]))
  magnitude = np.abs(np.array(fourier, copy=True))
  frequencies = np.fft.fftfreq(len(x[0]), 0.1)

  figure(figsize=(8, 6), dpi=80)

  plt.subplot(3, 1, 1)
  plt.plot(x[0], x[1], color ='blue')
  
  plt.xlabel('Time')
  plt.ylabel('Amplitude')
  plt.title("Signal")

  # plotting the magnitude spectrum of the signal 
  plt.subplot(3, 1, 2)
  plt.plot(frequencies, magnitude, color ='green')
    
  plt.title("Magnitude Spectrum of the Signal's Fourier Transform")

  shifted_frequencies = np.fft.fftshift(frequencies)
  phase = np.angle(np.fft.fftshift(fourier))


  plt.subplot(3, 1, 3)
  #, Fc = -0.5, color ='purple')
  plt.plot(shifted_frequencies, phase, color ='purple')
  plt.xlim(-0.25, 0.25)
    
  plt.title("Phase Spectrum of the Signal's Fourier Transform")

  plt.tight_layout()
  plt.show()

  domain_phase = sorted(zip(magnitude, phase, frequencies), key=lambda tup: tup[0], reverse=True)
  highest_frequencies = []
  for i in domain_phase:
    if i[1] >= 0:
      highest_frequencies.append(i[2])

  print("n highest magnitude point frequencies: " + str(highest_frequencies[0:n]))

  #return highest_magnitude_frequencies
```
